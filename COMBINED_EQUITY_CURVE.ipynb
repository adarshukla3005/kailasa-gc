```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

files_and_labels = {
    'datasets/combined_dataset/NBDS1equity.csv': 'NB Daily Strategy 1',
    'datasets/combined_dataset/NBFS2equity.csv': 'NB 15min Strategy 2',
    'datasets/combined_dataset/NBSS3equity.csv': 'NB 60min Strategy 3',
    'datasets/combined_dataset/NFDS1equity.csv': 'NF Daily Strategy 1',
    'datasets/combined_dataset/NFFS2equity.csv': "NF 15min Strategy 2",
    'datasets/combined_dataset/NFSS3equity.csv': 'NF 60min Strategy 3'
}

# Load all equity curves
equity_series = []
for file in files_and_labels:
    s = pd.read_csv(file).iloc[:, 0]
    equity_series.append(s)

# Get the max length (longest series)
max_len = max(len(s) for s in equity_series)

# Plot with stretched x-axis for each series
plt.figure(figsize=(12, 6))
colors = ['r', 'b', 'g', 'm', 'c', "gold"]

for i, (file, label) in enumerate(files_and_labels.items()):
    y = equity_series[i].values
    x = np.linspace(0, max_len - 1, len(y))  # Stretch x to match max length
    plt.plot(x, y, label=label, color=colors[i])

plt.title('Equity Curves (Raw Values with Proportional X-Axis)')
plt.xlabel('Proportional Time Index')
plt.ylabel('Equity Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
